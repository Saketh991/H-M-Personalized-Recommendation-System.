```python
import pandas as pd
```

```python
#Load the transactions data
transactions_df=pd.read_csv("transactions_train.csv")
```

```python
# Load the customers and articles data
customers_df=pd.read_csv("customers.csv")
articles_df=pd.read_csv("articles.csv")
```

```python
#Group by the data with the customer id column to get the transactions of each customer
temp = transactions_df.drop(['price','sales_channel_id','t_dat'],axis=1).groupby("customer_id")
```

```python
# Get the customer ids of the customers who bought atleast 70 products
filtered_customer_ids = temp.count()[temp.count()['article_id']>70].index
```

```python
# Filter the transactions by the customer ids who bought atleast 70 items
transactions_df = transactions_df[transactions_df['customer_id'].isin(filtered_customer_ids)]
```

```python
transactions_df
```

```python
# Filter the customers data with the ids who bought atleast 70 items
customers_df = customers_df[customers_df['customer_id'].isin(filtered_customer_ids)]
```

```python
# Filter the articles by the customer ids who bought atleast 70 items
articles_df = articles_df[articles_df['article_id'].isin(transactions_df.article_id.unique())]
```

```python
# Convert the t_dat column of the transaction to datetime format
transactions_df["t_dat"]=pd.to_datetime(transactions_df['t_dat'], infer_datetime_format=True)
```

```python
# Split the transactions data into train and test data by taking entire duration except last week as training data and the last week as test data
transactions_train=transactions_df[(transactions_df['t_dat'] <= '2020-09-15')]
```

```python
transactions_train
```

```python
transactions_test=transactions_df[(transactions_df['t_dat'] > '2020-09-15')]
transactions_test = transactions_test[transactions_test['customer_id'].isin(transactions_train["customer_id"])]
```

```python
transactions_test
```

```python
# To get only the customer id and the products they bought in the transactions test data
test=transactions_test.groupby("customer_id")
groups_test = dict(list(test))
test_data=pd.DataFrame(columns=["customer_id","products"])
for i in groups_test.keys():
  test_data=test_data.append({"customer_id":i,"products":groups_test[i]["article_id"].values},ignore_index=True)               
```

```python
test_data
```

```python
# Save all the subset data into the csv files 
customers_df.to_csv("customers_subset.csv",index=False)
articles_df.to_csv("articles_subset.csv",index=False)
test_data.to_csv("transactions_test_subset.csv",index=False)
transactions_train.to_csv("transactions_train_subset.csv",index=False)
```
